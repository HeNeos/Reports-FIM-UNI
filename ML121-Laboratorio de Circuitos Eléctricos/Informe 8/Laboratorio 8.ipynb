```python
import math
import matplotlib.pyplot as plt
```

```python
w = 377
v1 = [220.3, 210.3, 201, 190.1, 180.1, 170.1, 160, 150.2, 140, 130.3, 120]
v2 = [116, 110.9, 106, 100.3, 95, 89.8, 84.3, 79.2, 73.8, 68.8, 63.3]
i1 = [0.15, 0.13, 0.12, 0.1, 0.102, 0.094, 0.081, 0.076, 0.071, 0.064, 0.062]
w1 = [13.5, 12, 10.5, 9, 8.7, 7.5, 6, 6, 4.5, 3.3, 3]
```

```python
l1 = []
for i in range(0,len(v1)):
    l1.append((1/w)*((v1[i]/i1[i])**2 - (w1[i]/(i1[i]*i1[i]))**2)**(0.5))
r1 = []
for i in range(0,len(l1)):
    r1.append((w1[i]/(i1[i]*i1[i])))
m21 = []
for i in range(0,len(l1)):
    m21.append(1/w * v2[i]/i1[i])
```

```python
aux = []
indx = []
for i in range(0,1000):
    x = (60+i/(1000/60))
    aux.append(-0.0127*x+3.6992)
    indx.append(x)
plt.figure(figsize=(8,4))
plt.grid(True)
plt.xlabel('Volts')
plt.ylabel('Henrios')
plt.plot(v2,m21,'-go')
plt.plot(indx,aux,'--b')
plt.legend(('$M_{21}$ vs $V_{2}$','$y=-0.0127x+3.6992$'))
plt.savefig('fig1.png',dpi=300)
```
