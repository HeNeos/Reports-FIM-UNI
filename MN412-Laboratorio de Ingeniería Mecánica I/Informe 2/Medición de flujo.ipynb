```python
import math
import matplotlib.pyplot as plt
```

```python
alturat = [20.5,45,57.5,64]
alturaf = [0.4,0.49,0.5,0.75]
tiempo = [15,30.5,15.33,15.18]
volumen = [670,1954,1090,1160]
caudal = []
velocidads = []
for i in range(len(alturat)):
    alturat[i] = alturat[i]/100
    alturaf[i] = alturaf[i]*0.0254
    volumen[i] = volumen[i]*1e-6
    caudal.append(volumen[i]/tiempo[i])
    velocidads.append(math.sqrt(2*9.80665*alturat[i]))
print(caudal)
print(velocidads)
```

```python
import sklearn
import numpy as np
from sklearn.linear_model import LinearRegression
nvel = np.asarray(velocidads)
ncaudal = np.asarray(caudal)
ajli = np.polyfit(ncaudal,nvel,1)
for i in range(len(ncaudal)):
    nvel[i] = ncaudal[i]*ajli[0]+ajli[1]
```

```python
plt.figure(figsize=(10,6))
plt.plot(ncaudal,nvel,'-b')
plt.plot(caudal,velocidads,'ro')
plt.xlabel('Caudal')
plt.ylabel('Velocidad de salida')
plt.legend((r'Ajuste lineal ($V = 4.93187\cdot 10^{4}\cdot Q - 0.19019$)','Datos experimentales'))
plt.grid()
plt.savefig('velocidadcaudal.pdf')
```
