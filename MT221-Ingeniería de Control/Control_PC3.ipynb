```python
import control
from control.matlab import *
import numpy as np
import scipy as sc
import sympy as sy
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
def multiply_poly(a):
    p1 = np.array([1])
    for i in range(len(a)):
        polynom = np.array(a[i])
        p1 = sc.signal.convolve(p1, polynom)
    return p1
```

```python
multiply_poly([[1,0], [1, 5.6], [1, 10]])
```

```python
poles = []
zeros = []
n = 190
for k in range(1, n):
    p1 = np.array([98*k])
    p2 = np.array([1,12+1.4*k,98*k])
    p, z = control.pzmap(tf(p1,p2), Plot=False, grid=False)
    poles.append(p)
    zeros.append(z)

plt.figure(figsize=(8,6))

p_x = []
p_y = []
for i in range(0, n-1):
    p_x.append(np.real(poles[i][0]))
    p_y.append(np.imag(poles[i][0]))
plt.plot(p_x, p_y, '-b')
    
p_x = []
p_y = []
for i in range(0, n-1):
    p_x.append(np.real(poles[i][1]))
    p_y.append(np.imag(poles[i][1]))
plt.plot(p_x, p_y, '-b')
```
