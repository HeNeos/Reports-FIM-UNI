```python
import math
import matplotlib.pyplot as plt
```

```python
ang = [40,58,52]
pota = [12,11,11]
potr = []
fdp = []
for i in range(0,3):
    pota[i] = pota[i]*25
    ang[i] = ang[i]*math.pi/180
    fdp.append(math.cos(ang[i]))
    potr.append(pota[i]*math.tan(ang[i]))
```

```python
plt.figure(figsize=(8,5))
cond = [20.1,4.06,9.93]
pota.sort()
potr.sort()
cond.sort()
plt.plot(cond,pota,'-bo')
plt.plot(cond,potr,'-og')
plt.grid(True)
plt.xlabel('Capacitancia ($\mu$F)')
plt.ylabel('Potencia (W)')
plt.legend(('Potencia activa','Potencia reactiva'),shadow=True)
plt.savefig('Potencias.png',dpi=300)
```

```python
fdp.sort()
plt.figure(figsize=(6,4))
plt.plot(cond,fdp,'-ok')
plt.grid(True)
plt.xlabel('Capacitancia ($\mu$F)')
plt.ylabel('Factor de potencia')
plt.savefig('factordp.png',dpi=300)
```

```python
potexp = [50*13.4*1000/3600,50*13.1*1000/3600,50*13.84*1000/3600]
error = []
for i in range(0,3):
    error.append(100*abs(pota[i]-potexp[i])/potexp[i])
import pandas as pd
pd.DataFrame({'Condensador':cond,'Potencia teórica':pota,'Potencia experimental':potexp,'Error (%)':error},index=[1,2,3])
```
