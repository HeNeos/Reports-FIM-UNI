```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
from matplotlib import rc
```

```python
v = [0.074,0.88,1.623,1.9,1.94,2.393,2.433,2.847,2.952]
i = [1.12121,0.88,0.68279,0.59375,0.57396,0.43509,0.42684,0.30063,0.26934]
r = [0.066,1,2.377,3.2,3.38,5.5,5.7,9.47,10.96]
p = [0.083,0.7744,1.1082,1.1281,1.1135,1.0412,1.0385,0.8559,0.7951]
ax = plt.axes(projection='3d')
ax.plot_trisurf(r, i, v, linewidth=0.3, cmap = 'viridis')
plt.show()
```

```python
plt.ylabel(r'$P_{L}$ (mW)')
plt.xlabel(r'$R_{L}$ (k$\Omega$)')
plt.grid(True)
plt.plot(r, p, 'bo', r, p, 'k')
plt.savefig('ksken.png',dpi=300Ì£)
```

```python
i1 = []
for i in range(0,len(v)):
    i1.append(0.00510676*(r[i]*1000+3476.3771)/(r[i]*1000+3334.5118))
i1
```

```python
pf = []
for i in range(0,len(v)):
    pf.append(v[i]*i1[i]*1000)
pf
```

```python
n = []
for i in range(0,len(v)):
    n.append(p[i]/pf[i])
n
```

```python
plt.xlabel(r'$P_{f}$ (mW)')
plt.ylabel(r'$P_{L}$ (mW)')
plt.grid(True)
plt.plot(pf, p, 'bo', pf, p, 'k')
plt.savefig('vspotencia.pdf')
```

```python
plt.xlabel(r'$n$')
plt.ylabel(r'$R_{L}$ (k$\Omega$), Potencia(W)')
plt.grid(True)
plt.plot(n, r, '-bo')
plt.plot(n,p,'-bo',color='green')
plt.legend((r'$n$ vs $R_{L}$',r'$n$ vs Potencia'), shadow=True)
plt.savefig('vsnr.pdf')
```

```python
vth = []
rth = []
rl = []
il = []
for i in range(0,1000):
    vth.append(i+1)
    rth.append(i+1)
    rl.append(i+1)
    il.append(vth[i]/(rth[i]+rl[i]))
```

```python
pl = []
for i in range(0,1000):
    pl.append(il[i]*il[i]/rl[i])
```
