```python
import math
import matplotlib.pyplot as plt
```

```python
pelb = [2.6775,3.564,0.9768]
pela = [0.77,2.0223,2.904]
pmb = [0.5,1.18923,0.45683]
pma = [1.6825,2.243879,2.0457257]
indx = [1,2,3]
```

```python
plt.figure(figsize=(12,6))
plt.grid(True)
plt.plot(indx,pelb,'ob-')
plt.plot(indx,pela,'-ro')
plt.plot(indx,pmb,'-go')
plt.plot(indx,pma,'-mo')
plt.xlabel('Medida')
plt.ylabel('Potencia (kW)')
plt.legend(('Potencia eléctrica Baja presión','Potencia eléctrica Alta presión','Potencia mecánica Baja presión','Potencia mecánica Alta presión'))
plt.savefig('potencia1.pdf')
```

```python
for i in range(0,len(pma)):
    pma[i] = pma[i]/0.98
    pmb[i] = pmb[i]/0.98
```

```python
plt.figure(figsize=(12,6))
plt.grid(True)
plt.plot(indx,pelb,'ob-')
plt.plot(indx,pela,'-ro')
plt.plot(indx,pmb,'-go')
plt.plot(indx,pma,'-mo')
plt.xlabel('Medida')
plt.ylabel('Potencia (kW)')
plt.legend(('Potencia eléctrica Baja presión','Potencia eléctrica Alta presión','Potencia mecánica Baja presión','Potencia mecánica Alta presión'))
plt.savefig('potencia2.pdf')
```

```python
efb = []
efa = []
for i in range(0,len(pma)):
    efb.append(100*pmb[i]/pelb[i])
    efa.append(100*pma[i]/pela[i])
```

```python
plt.figure(figsize=(7,5))
plt.grid(True)
plt.plot(indx,efb,'ob-')
plt.plot(indx,efa,'og-')
plt.xlabel('Medida')
plt.ylabel('Eficiencia (%)')
plt.legend(('Eficiencia Baja presión','Eficiencia Alta presión'))
plt.savefig('eficiencia.pdf')
```
